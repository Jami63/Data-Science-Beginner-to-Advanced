```python
import matplotlib.pyplot as PLOT
```

```python
PLOT.hist(D["eco_score"] , bins = 40)
PLOT.xlabel("Eco Score")
PLOT.ylabel("Frequency")
PLOT.title("Eco Score Distribution")
PLOT.show()
```

```python
import seaborn as SE
```

```python
SE.kdeplot(D["eco_score"] , fill = True)
PLOT.xlabel("Eco Score")
PLOT.ylabel("Frequency")
PLOT.title("Eco Score Distribution")
PLOT.show()
```

```python
PLOT.scatter(D["eco_score"] , D["fuel_consumption"])
PLOT.xlabel("Eco Score")
PLOT.ylabel("Fuel Consumption")
PLOT.show ()
```

```python
PLOT.scatter(D["rpm_variation"] , D["eco_score"])
PLOT.xlabel("Eco Score")
PLOT.ylabel("RPM Variation")
PLOT.show ()
```

```python
PLOT.plot(D["eco_score"])
PLOT.xlabel("Eco Score")
PLOT.show ()
```

```python


C_ALL = ["rpm_variation","harsh_braking_count","idling_time",
        "fuel_consumption","acceleration_smoothness","eco_score"]

PLOT.figure(figsize=(10,5))
SE.boxplot(data=D[C_ALL])
PLOT.show()
```

```python
C_2 = ["rpm_variation" , "fuel_consumption"]
PLOT.figure(figsize=(10,5))
SE.violinplot(data=D[C_2])
PLOT.show()
```

```python
CORRELATION = D.corr ()
print(CORRELATION)
```

```python
PLOT.figure(figsize=(10,5))
SE.heatmap(CORRELATION , annot=True)
PLOT.show()
```

```python
SE.pairplot(D)
PLOT.show()
```
