```python
from sklearn.linear_model import LogisticRegression
```

```python
X = [[2] , [4] , [6] , [8]]
Y = [0 , 0 , 1 , 1]
```

```python
M = LogisticRegression()
```

```python
T = M.fit(X,Y)
```

```python
P = T.predict([[5]])
```

```python
if P == 1 :
  print("Pass")
else :
  print("Fail")
```

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
X = [[60] , [70] , [68] , [40]]
Y = [0 , 1 , 1 , 0]
```

```python
M = KNeighborsClassifier(n_neighbors=2)
```

```python
T = M.fit(X,Y)
```

```python
P = T.predict([[50]])
```

```python
if P == 1:
  print("Pass")
else :
  print("Fail")
```
